```python
import csv
import matplotlib.pyplot as plt

f=open('subwayfee.csv')
data=csv.reader(f)
next(data)
label=['유임승차','유임하차','무임승차','무임하차']
c=['skyblue','hotpink','green','yellow']
plt.rc('font',family='Malgun Gothic')    

for row in data:
    for i in range(4,8):
        row[i]=int(row[i])
    plt.figure(dpi=300)    
    plt.pie(row[4:8],labels=label,colors=c,autopct='%1.f%%')
    plt.axis('equal')
    plt.title(row[3]+' '+ row[1])
    plt.legend(loc='center right',bbox_to_anchor=(1,0,0.5,1))

    #plt.savefig('./images/'+row[3]+''+row[1]+'.png')
    plt.show()   
    break
```

## 유임승차 , 유임하차, 무임 승차 , 무임하차 비율을 piechart로 도식 후  images 폴더내에 저장

```python
import csv
import matplotlib.pyplot as plt
f= open('subwayfee.csv')
data=csv.reader(f)
next(data)

label= ['유임승차', '유임하차', '무임승차', '무임하차']
c=['skyblue','hotpink','green','yellow']

for row in data:
    for i in range(4,8):
        row[i] = int(row[i])


    plt.figure(dpi=300)
    plt.rc('font',family='Malgun Gothic')

    plt.title(row[3]+' '+row[1])
    plt.pie(row[4:8],labels=label,colors=c,autopct='%1.f%%')
    plt.axis('equal')
    plt.savefig('./images/'+row[3]+''+row[1]+'.png')
    plt.show()
```
