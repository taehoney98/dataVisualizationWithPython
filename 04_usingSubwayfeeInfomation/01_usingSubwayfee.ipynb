## 무임승차 대비 유임승차 비율높은 지하철역 출력
무임승차값 row[6] 대비 유임승차 값 row[4] 가 높은 지하철 역을 계산해 출력한다.

rate = 유임승차인원 / 무임승차인원

```python
import csv
f=open('subwayfee.csv')
data=csv.reader(f)
next(data)

maxRate=0
eachStationRate=0

for row in data:
    for i in range(4,8):
        row[i]=int(row[i])
    
    #무임승차 대비 유임승차 비율
    
    if row[6]!=0:# 0일경우 처리불가
        eachStationRate = row[4]/row[6]

        if eachStationRate> maxRate:
            maxRate=eachStationRate
            print("* update:" ,row,round(eachStationRate ,2))

#row[4] : 유임승차
#row[6] : 무임승차
```

## 유임승차 비율이 94%이상인 지하철 역 출력

100000명 이상의 탑승객이 존재하는 정거장 중 전체 탑승객 대비 유임승차 비율이 94%이상인 지하철 역과 통행량을 출력한다.

rate = 유임승차인원 / 유임+무임 전체 승차인원

```python
import csv
f=open('subwayfee.csv')
data=csv.reader(f)
next(data)

maxRate=0
eachStationRate=0
numOfPassengers=0
for row in data:
    for i in range(4,8):
        row[i]=int(row[i])
    
    numOfPassengers =row[4]+row[6]
    
    if row[6]!=0 and row[4]+row[6]>100000 :
        eachStationRate =row[4]/numOfPassengers  #전체 탑승객 대비 유임승차 비율
        if eachStationRate >maxRate:
            maxRate=eachStationRate
            max_station=row[1]+' '+row[3] +'역'
            #print("* update:" ,row,round(eachStationRate ,2))

        if eachStationRate >0.94:
            print("* station-over94%:" ,row,round(eachStationRate ,2))
            
print("가장 유임승차 비율이 높은 역: " , max_station , " 유임승차 비율: ", round(maxRate*100 ,2),"%")
```

## 유무임 승하차 인원이 가장 많은 지하철 역 찾기

```python
import csv
f= open('subwayfee.csv')
data=csv.reader(f)
next(data)
maxNumOfP = [0] *4 # maxNumberOfPassangers
max_station=[''] *4 # 통행량이 가장 많은 지하철역 +호선명 저장
label= ['유임승차', '유임하차', '무임승차', '무임하차']
for row in data:
    for i in range(4,8):
        row[i] = int(row[i])
        if row[i] >maxNumOfP [i-4]:
            maxNumOfP[i-4] =row[i]
            max_station[i-4] = row[3] +' ' +row[1]
            # row[3]: 지하철역 row[1]: 호선명

for i in range(4):
    print(label[i]+":"+ max_station[i] , maxNumOfP[i],"명")
```
