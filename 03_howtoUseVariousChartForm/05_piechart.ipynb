```python
import matplotlib.pyplot as plt
size = [2441,2312,1031,1233]#size 순서대로 나타냄 어떤게 어떤건지 파악 어려움
plt.axis('equal')
plt.pie(size)
plt.show()
```

```python
import matplotlib.pyplot as plt
size = [2441,2312,1031,1233]#size 순서대로 나타냄
plt.rc('font',family = 'Malgun Gothic')
label=['A형' , 'B형','O형','AB형']#label 적시
plt.axis('equal')
plt.pie(size , labels=label , autopct='% .1f%%') #비율 소수점 첫짜리까지
plt.legend(title = '혈액형' , loc ='center right' , bbox_to_anchor= (1,0,0.5,1))
plt.show()
```

```python
import matplotlib.pyplot as plt
size = [2441,2312,1031,1233]#size 순서대로 나타냄
plt.rc('font',family = 'Malgun Gothic')
label=['A형' , 'B형','O형','AB형']#label 적시
plt.axis('equal')
plt.pie(size , labels=label , autopct='% .1f%%' ,explode=(0,0,0.1,0)) #비율 소수점 첫째 , 돌출형
plt.legend(title = '혈액형' , loc ='center right' , bbox_to_anchor= (1,0,0.5,1) )
plt.show()
```

```python
import matplotlib.pyplot as plt
size = [2441,2312,1031,1233]#size 순서대로 나타냄
plt.rc('font',family = 'Malgun Gothic')
label=['A형' , 'B형','O형','AB형']
color = ['pink', 'skyblue','red','black']#list로 인자값 넘겨주는게 용의
explodeEff= [0,0,0.5,0]
plt.axis('equal')
plt.pie(size , labels=label , autopct='% .1f%%' ,colors=color,explode=explodeEff ) 
plt.legend(title = '혈액형' , loc ='center right' , bbox_to_anchor= (1,0,0.5,1) )
plt.show()
```
