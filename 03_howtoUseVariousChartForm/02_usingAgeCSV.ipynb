# usingAgeCSV.ipynb
행정안전부(www.mois.go.kr)에서 제공하는 주민등록 인구 통계 (jumin.mois.go.kr) 사이트에 접속한 후,

연령별 인구현황 메뉴에서 남·여 구분 없이 연령 구분 단위를 1세로 설정 후 인구 데이터를 다운받는다.
파일명을 age.csv로 수정 후 파이썬파일과 같은 폴더에 위치시켜 데이터를 시각화한다.

```python
import csv
f = open('age.csv')
data = csv.reader(f)
result = []

name = input("지역의 이름 입력하세요:")
for row in data : 
    if name in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(",","")))
            
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.title(name+' 지역의 인구구조')
plt.plot(result)
plt.show()
```

plot chart로 표현할 수도 있지만, 동일한 데이터를  bar chart 로 표현하는 경우 직관성이 향상된다.

```python
import csv
f = open('age.csv')
data = csv.reader(f)
result = []

name = input("지역의 이름 입력하세요:")
for row in data : 
    if name in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(",","")))



import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

plt.figure(figsize=(10 , 5))
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus']= False
plt.title( name+ ' 지역의 인구구조')
plt.xlabel("age")
plt.ylabel('cnt')
plt.plot(result)
plt.bar(range(101) , result)
plt.show()
```

bar chart 뿐만 아니라 수직형인 barh chart로도 표현할 수 있다.

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

plt.figure(figsize=(10 , 5))
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus']= False
plt.title(name+' 지역의 인구구조')
plt.xlabel("age")
plt.ylabel('cnt')

plt.barh(range(101) , result)
plt.show()
```
