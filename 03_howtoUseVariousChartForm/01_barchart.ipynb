```python
import matplotlib.pyplot as plt
#plt.bar(막대를 표시할 위치 , 막대의 높이)
plt.bar([0,1,2,4,6,10],[1,2,3,5,6,7])
plt.show()
```

```python
plt.bar([0,3,2,1,6,10],[1,2,3,4,5,6])
plt.show()
# 위치가 정렬되어 있지 않으면 정렬해 표시
```

```python
plt.bar(range(6),[1,2,3,4,5,6])
plt.show()
#range함수를 통해 정렬해 표현 
```

```python
import csv
f = open('age.csv')
data = csv.reader(f)

result =[]

for row in data:
    if '신도림' in row [0]:
        for i in row [3:]:
            result.append(int(i))
            
import matplotlib.pyplot as plt
plt.bar(range(101), result)
plt.show()
```

```python
plt.barh(range(101),result) 
#range(101)이 y축이 되고, result가 x축이 됨
```
