# 비슷한 인구구조를 가진 지역 찾기

age.csv파일에 저장된 데이터 중 row[2]에 저장된 2020년04월_계_연령구간인구수와 row[3]부터 저장된 특정 연령대의 인구수 정보를 이용해  
총인구수 대비 특정구간의 인구수의 비율을 계산해 입력받은 지역의 인구구조와 유사한 형태의 인구구조를 가진 지역명을 도출한다.  

```python
import numpy as np
import csv

f=open('age.csv')
data= csv.reader(f)
next(data)
data = list(data)
mn=1
result_name=''
result=0
name =input("지역명 입력")
'''
row[2] 2020년04월_계_연령구간인구수
'''





for row in data:
    for i in range(1,len(row)):
        row[i]=row[i].replace(",","")
    
    
    if name in row[0]:
        home=np.array(row[3:], dtype=int) /int(row[2].replace(",",""))


```

```python
for row in data:
    
    away = np.array(row[3:] ,dtype=int)/int(row[2].replace(",",""))
    
    s=np.sum((home-away)**2)
    if s<mn and name not in row[0]:
        mn =s
        result_name =row[0]
        result= away
        

```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.rc('font', family='Malgun Gothic')
plt.title(name + '지역과 가장 비슷한 인구구조를 가진 지역')
plt.plot(home , label=name)
plt.plot(result,  label =result_name)
plt.legend()
plt.show(0)
```
