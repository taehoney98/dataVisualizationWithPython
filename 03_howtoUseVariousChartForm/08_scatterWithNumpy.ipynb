# scatterWithNumpy.ipynb

x,y에 random함수를 통해 저장한 데이터를 scatter 함수를 통해 표현한다.

```python
import matplotlib.pyplot as plt
import random
x=[]
y=[]
size=[]#empty list 생성
for i in range(100):
    x.append(random.randint(50,100))#50~100 사이의 랜덤정수 x에 더함
    y.append(random.randint(50,100))
    size.append(random.randint(10,100))#점의 크기는 10부터 100까지
    
plt.scatter(x,y ,s=size)
plt.show()
```

```python
plt.scatter(x,y,s=size,c=size,cmap='jet',alpha =0.5)
#정석은 color 값으로 개수를 주어야하지만 리스트의 개수로 파악해서 이것도 ok
#alpha 값을 지정해 겹치는 부분 투명도로 표현
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x= np.random.randint(10,100 ,200)
y = np.random.randint(10,100 ,200)
size= np.random.randint(10,100 ,200)
    
plt.scatter(x,y,s=size,c=x,cmap='jet',alpha=0.7) #alpha 투명도 0.7
plt.colorbar()
plt.show()
```
