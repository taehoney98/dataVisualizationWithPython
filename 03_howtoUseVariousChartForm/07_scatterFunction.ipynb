# scatterFunction.ipynb

scatter 함수를 통해 좌표계 상의 점들을 산점도로 표현한다.

```python
import matplotlib.pyplot as plt
plt.scatter([1,2,3,4] ,[10,20,30,40])
plt.show()
```

```python
import matplotlib.pyplot as plt
size=[100,200,300,400]
plt.scatter([1,2,3,4] ,[10,20,30,40] , s=size)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter([1,2,3,4], [10,20,30,40],s=[100,2000,250,300])
# 좌표 (1,10) (2,20) (3,30) (4,40)
# s 속성 사이즈 100 200 250 300
plt.show()
```

```python

plt.scatter([1,2,3,4], [10,20,30,40],s=[100,200,250,300], c=['red' , 'blue' , 'green', 'gold'])
# 좌표 (1,10) (2,20) (3,30) (4,40)
# s 속성 사이즈 100 200 250 300
# c 컬러 점의 개수만큼 지정
plt.show()
```

```python

plt.scatter([1,2,3,4], [10,20,30,40],s=[100,2000,250,300], c=range(4))
plt.colorbar()# 0~3까지의 각각 색상지정
plt.show()
#color bar로 표현
```

```python
plt.scatter([1,2,3,4] ,[10,20,30,40],s=[30,60,90,120], c= range(4), cmap='jet')
# jet 라는 컬러맵 사용
# matplot에는 jet 이외에도 rainbow, prism등 여러 컬러맵 탑재되어 있음.
plt.colorbar()
plt.show()
#scatter camp 속성
# color map 은 RGBA(red green blue alphaf를 0~255까지 투명도로 출력)
# name 컬러맵의 이름 'jet'지정
# int RGB를 (Pulse code modulation의) 양자화된 수로 지정
```

```python
size=[100,200,300,400]
color=['red', 'blue','pink','purple']
plt.scatter([1,2,3,4],[10,20,30,40], s=size, c=range(4),cmap='jet')

plt.colorbar(orientation='horizontal',fraction=0.1)
#컬러바를 가로로 지정
plt.show()
```
