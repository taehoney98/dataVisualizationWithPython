matplotlib의 plot을 통해 내 생일인 05월 18일의 최고기온과 최저기온의 변화를 시각화한다.

```python
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1]!= '':
        result.append(float(row[-1])) #최고기온 값 추가
print(len(result)) 
```

```python
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data=csv.reader(f)
next(data)
high=[]
low=[]

year= []
for i in range(1920 ,2020): # 1920년 부터 2019년 까지 year에 append
    year.append(i)
for row in data:
    if row[-1] !='' and row[-2] != '':
        if 1920 <= int (row[0].split('-')[0]): # 년도가 1920년 이상이면
            if row[0].split('-')[1] == '05' and row[0].split('-')[2]=='18' :
                high.append(float(row[-1]))# 최고기온
                low.append(float(row[-2])) # 최저기온

#set chart size: 폭 10 인치 높이 2인치
#plt.figure(figsize= (10 , 2))
plt.rc('font', family ='Malgun Gothic')
plt.rcParams['axes.unicode_minus']= False
plt.title('1920-2019 100년간 내생일의 기온 변화 그래프')

plt.plot(high,color='red' , label='high_temp')
plt.plot(low, color ='skyblue' ,label='low_temp')
plt.legend()
plt.show()       

f.close()
```

## birthday입력으로 평균기온 출력
input()문을 통해 birthday를 입력받아 해당  날짜의 평균기온을 출력한다.

```python
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data=csv.reader(f)
next(data)
avg=[]
year= []

birthday=input('생일을 입력하세요: ')
birthDay=birthday.split(' ')

for i in range(1920 ,2020): 
    year.append(i)
for row in data:
    if row[2] != '':
        if 1920 <= int (row[0].split('-')[0]): 
            if row[0].split('-')[1] == birthDay[0] and row[0].split('-')[2]==birthDay[1] :
                avg.append(float(row[2]))


plt.rc('font', family ='Malgun Gothic')
plt.rcParams['axes.unicode_minus']= False
plt.title(f'1920-2019 100년간 {birthDay[0]}월 {birthDay[1]}일의 평균기온 변화 그래프')

plt.plot(avg ,color='green' , label='avg_temp')
plt.legend()
plt.show()       

f.close()
```
