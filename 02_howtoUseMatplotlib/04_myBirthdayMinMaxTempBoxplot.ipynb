앞서 plot으로 표현한 05월 18일의 기온 변화 그래프를 boxplot을 통해 표현한다.

```python
import csv
import matplotlib.pyplot as plt  

f = open('seoul.csv')
data=csv.reader(f)
next(data)
high=[]
low=[]

for row in data:
    if row[-1] !='' and row[-2] != '':
        date = row[0].split('-') #date[0]= year / date[1] =month / date[2] = day
        if 1920 <= int (date[0]): # 년도가 1920년 이상이면
            if date[1] == '05' and date[2]=='18' :
                high.append(float(row[-1]))
                low.append(float(row[-2]))


plt.rc('font', family ='Malgun Gothic') #맑은 고딕 기본 글꼴 설정
plt.rcParams['axes.unicode_minus']= False #마이너스 기호 깨짐 방지
plt.title('1920-2019 100년간 내 생일의 최저온도, 최고온도')
plt.xlabel('생일: 5월 18일')          
plt.ylabel('기온')
plt.plot(high , 'hotpink' , label = 'high')
plt.plot(low , 'skyblue' ,label ='low')
plt.legend()
plt.show() 


f.close()
```

## Box plot


1. 최소값
2. 25%를 의미하는 제 1사분위(Q1, Quartiles1) 
3. 중앙값을 의미하는 제 2사분위(Q2, Quartiles2) 
4. 75%를 의미하는 제 3사분위(Q3, Quartiles3) 
5. 최대값
자료의 5가지 값을 box plot으로 나타낸다.

```python
import csv
import matplotlib.pyplot as plt  

f = open('seoul.csv')
data=csv.reader(f)
next(data)
high=[]
low=[]

for row in data:
    if row[-1] !='' and row[-2] != '':
        if 1920 <= int (row[0].split('-')[0]): # 년도가 1920년 이상이면
            if row[0].split('-')[1] == '05' and row[0].split('-')[2]=='18' :
                high.append(float(row[-1]))
                low.append(float(row[-2]))


plt.rc('font', family ='Malgun Gothic')
plt.rcParams['axes.unicode_minus']= False
plt.title('1920-2019 100년간 내 생일의 최저온도, 최고온도')
plt.xlabel('생일: 5월 18일')          
plt.ylabel('기온')
#plt.boxplot([low,high])    기본 출력코드
#plt.boxplot([low,high] , showfliers = False) 이상치 삭제코드
red_square = dict(markerfacecolor= 'r' , marker='s')
plt.boxplot([low,high] , flierprops=red_square )    #이상치 마커 사용
plt.show() 

f.close()
```

## 1월,8월 최고기온을 boxplot으로 표시

seoul.csv 의 row[-1]에 최고기온이 저장되어 있다는 정보를 이용해 1월과 8월의 최고기온의 분포를 boxplot을 통해 나타낸다.

```python
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug=[]
jan=[]

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month =="01":
            jan.append(float(row[-1]))
        
        
        if month =="08":
            aug.append(float(row[-1]))
            
            
import matplotlib.pyplot as plt

plt.boxplot([aug,jan])
plt.show()


f.close()
```

## 월별 평균기온 boxplot으로 표시
seoul.csv의 row[2]가 평균기온을 나타낸다는 정보를 이용해 1월~12월의 매년 각각의 평균기온의 분포를 boxplot으로 나타낸다.

```python
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)
month=[[],[],[],[],[],[],[],[],[],[],[],[]] #month[0] == 1월용 empty list ~   month[11] == 12월용 empty list

for row in data:
    if row[-1] != '':
        mon=row[0].split('-')[1]
        month[int(mon)-1].append(float(row[2]))
        
plt.boxplot(month)
plt.show()


f.close()
```

## 5월의 평균기온의 분포 boxplot으로 표시

같은 방식으로 31개의 빈 리스트를 만들고, row[0].split('-')[1]이 월을 의미하는 점을 이용해 
5월의 평균기온의 분포를 boxplot으로 표시한다.

```python
import matplotlib.pyplot as plt
import csv

f=open('seoul.csv')
data = csv.reader(f)
next(data)

day=[]
for i in range (31):
    day.append([])

for row in data:
    if row [-1]!='':
        if row[0].split('-')[1] =='05':
            day[int(row[0].split('-')[2])-1].append(float(row[2]))
            
            
%matplotlib inline
plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.boxplot(day,showfliers=False)

f.close()
```
