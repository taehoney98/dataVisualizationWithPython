# subwayTimeStationMaxMin.ipynb
2020_04_subwaytime.csv 파일을 이용해 시간대 별 최대 승차인원과 최대 하차인원을 도출하고, 해당 그래프 값을 images 폴더에 저장한다. 


```python
import csv
f=open('2020_04_subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)

maxOfP=[0] *24#maxof passager
max_station=[""]*24# maxPassager Station

for row in data:
    row[4:]=map(int,row[4:])
    
    for j in range(24):
        a = row[j*2+4]
        if a >maxOfP[j]:
            maxOfP[j]= a
            max_station[j]=row[3] + '('+str(j+4)+'시)'
            
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.bar(range(24), maxOfP)
plt.xticks(range(24), max_station , rotation=90)
plt.show()
```

```python
import csv
f=open('2020_04_subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)

maxOfP=[0] *24#maxof passager
max_station=[""]*24# maxPassager Station

for row in data:
    row[4:]=map(int,row[4:])
    
    for j in range(24):
        b = row[j*2+5]
        if b >maxOfP[j]:
            maxOfP[j]= b
            max_station[j]=row[3] + '('+str(j+4)+'시)'
            
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.bar(range(24), maxOfP , color='r')
plt.xticks(range(24), max_station , rotation=90)
plt.show()
```

```python
import csv
f=open('2020_04_subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)

maxOfP=[0] *24#maxof passager
maxS=[""]*24# maxPassager Station
offMaxP=[0] *24#offMax of passager
offMaxS=[""]*24# offMax of passager Station
temptime=0

for row in data:
    row[4:]=map(int,row[4:])
    
    for j in range(24):
        temptime=int(str(j+4))
        if temptime> 24:
            temptime=temptime-24
        tempP=row[j*2+4]    
        if tempP>maxOfP[j]:
            maxOfP[j]=tempP
            maxS[j]=row[3]+"("+row[1]+")"+"("+str(temptime)+")"
        tempP=row[j*2+4+1]
        if tempP>offMaxP[j]:
            offMaxP[j]=tempP
            offMaxS[j]=row[3]+"("+row[1]+")"+"("+str(temptime)+")"
            
import matplotlib.pyplot as plt

plt.figure(figsize=(10,3))
plt.rc('font',family='Malgun Gothic')
plt.title('2020년 4월 시간대별로 최대 승차인원')

plt.bar(range(24),maxOfP,color='red',label='승차')
plt.xticks(range(24),maxS,rotation=90)
plt.savefig('images/'+'시간대별로 최대 승차인원'+'.png',bbox_inches='tight')    


plt.figure(figsize=(10,3))
plt.rc('font',family='Malgun Gothic')
plt.title('2020년 4월 시간대별로 최대 하차인원')

plt.bar(range(24),offMaxP,color='blue',label='승차')
plt.xticks(range(24),offMaxS,rotation=90)    
plt.savefig('images/'+'시간대별로 최대 하차인원'+'.png',bbox_inches='tight') 

plt.show()    
  
```

```python
import csv
import matplotlib.pyplot as plt
f=open('2020_04_subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)
s_in=[0]*24 # 승차인원 저장 리스트
s_out=[0]*24 # 하차인원 저장 리스튼


for row in data:
    row[4:]=map(int,row[4:])
    for i in range(24):
        s_in[i]+=row[4+i*2]
        s_out[i]+=row[5+i*2]



plt.rc('font',family='Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in,label='승차')
plt.plot(s_out,label='하차')
plt.legend()
plt.show()

#1e7 1* 10^7 천만명 단위라는 것을 나타냄
```
